```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pymc3 as pm
import theano.tensor as tt
```

```python
plt.rcParams['figure.figsize'] = 10, 10
```

```python
df_sleep_data = pd.read_csv('data/sleep_data.csv', index_col=0)
```

```python
list_sleep_labels = ['9:00', '9:30', '10:00', '10:30', '11:00', '11:30', '12:00']
```

```python
plt.scatter(df_sleep_data['time_offset'], df_sleep_data['indicator'], s= 60, alpha=0.01, facecolor = 'b', edgecolors='b')
plt.yticks([0, 1], ['Awake', 'Asleep'])
plt.xlabel('PM Time')
plt.title('Falling Asleep Data', size = 18)
plt.xticks([-60, -30, 0, 30, 60, 90, 120], list_sleep_labels)
plt.show()
```

- We use Markov Chain Monte Carlo to approximate the logistic function to represent this behaviour

$$
P(\text{Asleep | Time}) = \frac{1}{1 + e^{\beta t + \alpha}}
$$

- To be able to draw samples, we need to create a prior distribution for $\beta$ and $\alpha$
    - i.e. we need to guess what the distributions should be

- The basic procedure for Markov Chain Monte Carlo in our problem is as follows:
    1. Select an initial set of values for alpha and beta, the parameters of the logistic function.
    2. Randomly assign new values to alpha and beta based on the current state.
    3. Check if the new random values agree with the observations. If they do not, reject the values and return to the previous state. If they do, accept the values as the new current state.
    4. Repeat steps 2 and 3 for the specified number of iterations.

- The algorithm returns all of the values it generates for alpha and beta
    - We can then use the average of these values as the most likely final values for alpha and beta in the logistic function.

```python
df_sleep_data.sort_values('time_offset', inplace=True)

array_time = df_sleep_data['time_offset'].values
array_observations = df_sleep_data['indicator'].values
```

- **Note**: in Bayesian software, it is common to use $\tau$ instead of $\sigma$
    - $\tau = \frac{1}{\sigma}$
    - Unlike $\sigma$, the higher the $\tau$ value, the **smaller** the dispersion of the data

```python
N_SAMPLES = 5000

with pm.Model() as sleep_model:
    # Create the alpha and beta parameters
    alpha = pm.Normal('alpha', mu=0.0, tau=0.01, testval=0.0)
    beta = pm.Normal('beta', mu=0.0, tau=0.01, testval=0.0)
    
    # Create the probability from the logistic function
    p = pm.Deterministic('p', 1. / (1. + tt.exp(beta * array_time + alpha)))
    
    # Create the bernoulli parameter which uses the observed dat
    observed = pm.Bernoulli('obs', p, observed=array_observations)

    # Using Metropolis Hastings Sampling
    step = pm.Metropolis()
    
    # Sample from the posterior using the sampling method
    sleep_trace = pm.sample(N_SAMPLES, step=step)
```

- We take the mean values for alpha and beta for our model

```python
alpha_est = np.mean(sleep_trace['alpha'])
beta_est = np.mean(sleep_trace['beta'])
```

- Now we create our logistic function

```python
def logistic(x, beta, alpha=0):
    return 1.0 / (1.0 + np.exp(np.dot(beta, x) + alpha))
```

```python
array_time_est = np.arange(np.min(array_time)- 15, np.max(array_time) + 15, 0.001)
```

```python
array_estimated_sleep = logistic(array_time_est, beta_est, alpha_est)
```

```python
plt.plot(array_time_est, array_estimated_sleep, color = 'navy', lw=3, label="Logistic Model")
plt.scatter(array_time, array_observations, edgecolor = 'slateblue', s=50, alpha=0.2, label='obs')
plt.title('Probability Distribution for Sleep with %d Samples' % N_SAMPLES);
plt.legend(prop={'size':18})
plt.ylabel('Probability')
plt.xlabel('PM Time');
plt.xticks([-60, -30, 0, 30, 60, 90, 120], list_sleep_labels);
```
