# Problem

- You have two medicines
    - Medicine 1: Combine 3 parts herb A, and 2 parts herb B
    - Medicine 2: Combine 4 parts herb A, and 1 part herb B
- Medicine 1 improves a person's health by 25 pts
    - Medicine 2 by 20 pts
    
- We have 25 units of herb A, and 10 units of herb B at our disposal
    - *How many units of each type of medicine should we make to maximize the number of pts?*

_____

# Solution

- First, we can loop through the possible combinations and see the number of pts

```python
max_pts = 0

for n_units_A in range(26):
    for n_units_B in range(11):
        A_1, B_1 = n_units_A, n_units_B
        A_2, B_2 = 25 - A_1, 10 - B_1
        
        total_pts_1 = 0
        while A_1 >= 3 and B_1 >= 2:
            A_1 -= 3
            B_1 -=2
            total_pts_1 += 25
            
        total_pts_2 = 0
        while A_2 >= 4 and B_2 >= 1:
            A_2 -= 4
            B_2 -=1
            total_pts_2 += 20
            
        total = total_pts_1 + total_pts_2
        if total > max_pts:
            max_pts = total            
            print('Medicine 1: n_units_A = {}, n_units_B = {}, total_pts_1 = {}'.format(n_units_A, n_units_B, total_pts_1))
            print('Medicine 2: n_units_A = {}, n_units_B = {}, total_pts_2 = {}'.format(25-n_units_A, 10-n_units_B, total_pts_2))
            print('Total Pts: {}'.format(total_pts_1 + total_pts_2))
            print('\n')
```

- After looping through the different combinations, it looks like the optimal allocation uses:
    - 9 units of herb A and 6 of herb B for Medicine 1
    - 16 units of herb A and 4 of herb B for Medicine 2
        - The total number of points produced is equal to 155

- We can visualize this finding

```python
def medicine_1(A_1, B_1):
    total_pts_1 = 0
    while A_1 >= 3 and B_1 >= 2:
        A_1 -= 3
        B_1 -=2
        total_pts_1 += 25
    return total_pts_1
```

```python
def medicine_2(A_2, B_2):
    total_pts_2 = 0
    while A_2 >= 4 and B_2 >= 1:
        A_2 -= 4
        B_2 -=1
        total_pts_2 += 20
    return total_pts_2
```

```python
import pandas as pd
import numpy as np
```

```python
array_A1 = np.arange(26)
array_B1 = np.arange(11)
```

```python
A_1, B_1 = np.meshgrid(array_A1, array_B1)
grid = np.c_[A_1.ravel(), B_1.ravel()]
```

```python
df = pd.DataFrame(grid, columns=['A_1', 'B_1'])
df['A_2'] = 25 - df['A_1']
df['B_2'] = 10 - df['B_1']
```

```python
df['pts_1'] = df.apply(lambda x: medicine_1(x['A_1'], x['B_1']), axis=1)
```

```python
df['pts_2'] = df.apply(lambda x: medicine_2(x['A_2'], x['B_2']), axis=1)
```

```python
df['total_pts'] = df['pts_1'] + df['pts_2']
```
