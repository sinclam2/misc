```python
from bs4 import BeautifulSoup
from urllib.request import urlopen, urlretrieve
import shutil
import re
import os
```

- This is a pretty hacky way of doing it, but fuck it

```python
list_links = []

html_page = urlopen("http://feeds.tvo.org/podcasts/video/AllanGreggInConversation")
soup = BeautifulSoup(html_page)
for link in soup.find_all('guid'):
    str_link = re.search('<guid ispermalink="false">(.*)</guid>', str(link)).group(1)
    list_links.append(str_link)
```

```python
str_page = str(soup)

list_titles = [re.search('<title>(.*)</title>', str(x)).group(1) for x in soup.find_all('title')][2:]

for i, str_link in enumerate(list_links):
    str_title = list_titles[i].replace(',','').replace('.','').replace(' ','_').replace('"','').replace('?','')
    file_extension = str_link[-4:]
    new_title = 'videos/' + str_title + file_extension   
    if str_title + file_extension not in os.listdir('videos'):
        urlretrieve(str_link, new_title)
    print('Successfully saved {}'.format(new_title))
```
