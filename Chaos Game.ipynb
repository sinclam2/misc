```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 10
```

```python
dict_origin = {0: (0,0),
              1: (1,0),
              2: (0.5,1)}
```

```python
def half_distance(x_0, y_0, corner):
    origin_x, origin_y = dict_origin[corner]
    h = np.sqrt((x_0-origin_x)**2+(y_0+0-origin_y)**2)
    if corner == 0:
        sin_theta = y_0/h
        y_1 = (h/2)*sin_theta
        cos_theta = x_0/h
        x_1 = (h/2)*cos_theta
    elif corner == 1:
        sin_theta = y_0/h
        y_1 = (h/2)*sin_theta
        cos_theta = (1-x_0)/h
        x_1 = 1-(h/2)*cos_theta 
    elif corner == 2:
        sin_theta = (x_0-0.5)/h
        x_1 = 1-(h/2)*sin_theta
        cos_theta = (1-y_0)/h
        y_1 = 1-(h/2)*cos_theta
        
    return x_1, y_1
```

```python
starting_point_x = np.random.random()
starting_point_y = np.random.random()

x = [starting_point_x]
y = [starting_point_y]

x_0 = starting_point_x
y_0 = starting_point_y

for i in range(100000):
    corner = np.random.randint(3)
    x_new, y_new = half_distance(x_0, y_0, corner)
    x.append(x_new)
    y.append(y_new)
    x_0 = x_new
    y_0 = y_new
```

```python
plt.scatter(x,y, s  = 0.1)
plt.show()
```
