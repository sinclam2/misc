# Liquidity Ratios

- These ratios measure how easily a company will be able to meet its short-term obligations
    - Higher is usually better

## Current Ratio

- Most common liquidity ratio

$$
\text{Current Ratio} = \frac{\text{Current Assets}}{\text{Current Liabilities}}
$$

- **Recall**:
    - *Current Assets* means cash and other assets that can quickly be converted to cash (i.e. within a year)
        - In other words, the liquid assets
    - *Current Liabilities* means obligations that need to be paid within a year
    
- Putting it all together, this means that **if the current ratio is greater than 1, it means the business could pay this year's obligations without worrying too much**

## Quick Ratio

- Similar to the current ratio
    - *Main Difference*: The quick ratio excludes inventory
    
$$
\text{Quick Ratio} = \frac{\text{Current Assets} - \text{Inventory}}{\text{Current Liabilities}}
$$

- The reason why we remove inventory from the current assets is because **it gives a worst-case-scenario assessment**
    - *Can the company meet this years obligations even if they don't make any more sales?*
        - This means that **if the quick ratio is above 1, the business can definitely meet this year's obligations without worrying too much**

### Example

**Assets**

- Cash and Cash Equivalents: 40
- Inventory: 100
- Accounts Receivable: 60
- PP & E: 300
- **Total**: 500

**Liabilities**

- Accounts Payable: 50
- Income Tax Payable: 150
- **Total**: 200

**Owners' Equity**

- Common Stock: 160
- Retained Earnings: 140
- **Total**: 300


- **Total Liabilities + Owners' Equity**: 500

$$
\text{Current Ratio} = \frac{40 + 100 + 60}{50 + 150} = \frac{200}{200} = 1
$$

$$
\text{Quick Ratio} = \frac{40 + 60}{200} = \frac{100}{200} = 0.5
$$

- *What do these ratios tell us?*
    1. Since the current ratio is greater than 1, this means that the business should be able to meet its current liabilities
    2. Since the quick ratio os 0.5, this means the business **needs to maintain sales to be able to meet its short-term obligations**

# Profitability Ratios

- Net income can tell us a lot about a business, but it makes it hard to compare the effectiveness of different businesses
    - We need to think of net income with respect to other values (i.e. as a ratio/proportion) to give additional colour
    
## Return on Assets

- This tells us how profitable a company is *with respect to its size*
    - Here, the *size* is measured by total assets
    
$$
\text{Return on Assets} = \frac{\text{Net Income}}{\text{Total Assets}}
$$

- From this ratio, we're trying to answer: **how efficiently is this company generating profits using its assets?**


## Return on Equity

- This tells us how profitable a company is *with respect to its equity*

$$
\text{Return on Equity} = \frac{\text{Net Income}}{\text{Shareholders' Equity}}
$$

- From this ratio, we're trying to answer: **how efficiently is this company generating profits using its investors' money?**

- *Using these two metrics, we can compare the profitability of companies of completely different sizes*

### Example

**Revenue**

- Sales: 300
- COGS: -100
- **Gross Profit**: 200

**Expenses**

- Rent: 30
- Salaries and Wages: 80
- **Total Expenses**: 110


- **Net Income**: 90


- Using the balance sheet values from the previous example:

$$
\text{Return on Assets} = \frac{90}{500} = 0.18
$$

$$
\text{Return on Equity} = \frac{90}{300} = 0.3
$$

## Gross Profit Margin

- This tells us the proportion of sales that exceed their cost
    - The extra amount is used to cover additional overhead costs (E.g. rent, salaries, etc.)
    
$$
\text{Gross Profit Margin} = \frac{\text{Sales} - \text{Cost of Goods Sold}}{\text{Sales}}
$$

### Example

- Total Sales: 80
- COGS: 20

$$
\text{Gross Profit Margin} = \frac{80 - 20}{20} = \frac{60}{80} = 0.75
$$

- This tells us that for every sale, 75% of the revenue can be used to pay the costs of running the business

- **Note**: gross profit margins can't be compared across different industries
    - E.g. Retail is a narrow margin business, while legal services is a wide margin business
- We can, however, use gross profit margin to **compare businesses within the same industry**

# Financial Leverage Ratios

- These ratios tell us how a business has used debt (instead of equity) to finance its operations

## Debt Ratio

- This tells us **how much of a company's assets have been paid for by debt**

$$
\text{Debt Ratio} = \frac{\text{Liabilities}}{\text{Assets}}
$$

## Debt-to-Equity Ratio

- This tells us a company's **relative preference between debt and equity**

$$
\text{Debt-to-Equity Ratio} = \frac{\text{Liabilities}}{\text{Owners' Equity}}
$$

**Dangers of Financial Leverage**

- More leverage $\implies$ higher return on equity
- More leverage $\implies$ higher risk of bankruptcy


### Example

**Company ABC**

- Assets: 200
- Liabilities: 100
- Equity: 100

- Net Income: 15

$$
\text{Return on Equity} = \frac{15}{100} = 15\%
$$

**Company XYZ**

- Assets: 200
- Liabilities: 150
- Equity: 50

- Net Income: 15

$$
\text{Return on Equity} = \frac{15}{50} = 30\%
$$

- Both companies have the same total assets
    - But since **Company XYZ is more highly indebted, it gave its investors twice the level of return**
