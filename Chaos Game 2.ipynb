```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 10
```

```python
dict_origin = {0: (0,0),
              1: (1,0),
              2: (0.5,1)}
```

```python
def half_distance(x_0, y_0, corner):
    origin_x, origin_y = dict_origin[corner]

    x_mean = np.mean([x_0, origin_x])
    y_mean = np.mean([y_0, origin_y])
        
    return x_mean, y_mean
```

```python
starting_point_x = np.random.random()
starting_point_y = np.random.random()

x = [starting_point_x]
y = [starting_point_y]

x_0 = starting_point_x
y_0 = starting_point_y

for i in range(100000):
    corner = np.random.randint(3)
    x_new, y_new = half_distance(x_0, y_0, corner)
    x.append(x_new)
    y.append(y_new)
    x_0 = x_new
    y_0 = y_new
```

```python
plt.scatter(x,y, s  = 0.1)
plt.show()
```

**Doubling it (so it goes 3/4 of the way)**

```python
starting_point_x = np.random.random()
starting_point_y = np.random.random()

x = [starting_point_x]
y = [starting_point_y]

x_0 = starting_point_x
y_0 = starting_point_y

for i in range(100000):
    corner = np.random.randint(3)
    x_new, y_new = half_distance(x_0, y_0, corner)
    x_new, y_new = half_distance(x_new, y_new, corner)
    x.append(x_new)
    y.append(y_new)
    x_0 = x_new
    y_0 = y_new
    
plt.scatter(x,y, s  = 0.1)
plt.show()
```

**Now it'll go halfway toward the random corner, then halfway toward to corner +1**

```python
starting_point_x = np.random.random()
starting_point_y = np.random.random()

x = [starting_point_x]
y = [starting_point_y]

x_0 = starting_point_x
y_0 = starting_point_y

for i in range(100000):
    corner = np.random.randint(3)
    corner2 = (corner+1)%3
    x_new, y_new = half_distance(x_0, y_0, corner)
    x.append(x_new)
    y.append(y_new)
    x_new, y_new = half_distance(x_new, y_new, corner2)
    x.append(x_new)
    y.append(y_new)
    x_0 = x_new
    y_0 = y_new
plt.scatter(x,y, s  = 0.1)
plt.show()
```

**Now taking the mean of (x0, y0) and (x1, y1)**

```python
starting_point_x = np.random.random()
starting_point_y = np.random.random()

x = [starting_point_x]
y = [starting_point_y]

x_0 = starting_point_x
y_0 = starting_point_y

for i in range(100000):
    corner = np.random.randint(3)
    x_new, y_new = half_distance(x_0, y_0, corner)
    x_new = (x_0+x_new)/2.0
    y_new = (y_0+y_new)/2.0
    x.append(x_new)
    y.append(y_new)
    x_0 = x_new
    y_0 = y_new
plt.scatter(x,y, s  = 0.1)
plt.show()
```

# New Game: there are four vertices, but can't use the same one twice in a row

```python
dict_origin = {0: (0,0),
              1: (1,0),
              2: (1,1),
              3: (0,1)}
```

```python
starting_point_x = np.random.random()
starting_point_y = np.random.random()

x = [starting_point_x]
y = [starting_point_y]

x_0 = starting_point_x
y_0 = starting_point_y

for i in range(100000):
    if i == 0:
        corner = np.random.randint(4)
    else:
        corner = np.random.choice([j for j in range(4) if j != corner])
    x_new, y_new = half_distance(x_0, y_0, corner)
    x.append(x_new)
    y.append(y_new)
    x_0 = x_new
    y_0 = y_new
plt.scatter(x,y, s  = 0.1)
plt.show()
```
